```python
import pandas as pd

heatmap_df = pd.read_csv(
    "../data/gen_2/processed/fire_coordinate_weights.csv", dtype={"month": str}
)

heatmap_df = heatmap_df[heatmap_df["month"] == "01"]
heatmap_df["visible"] = 1
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    go.Densitymapbox(
        lat=heatmap_df.latitude,
        lon=heatmap_df.longitude,
        z=heatmap_df.visible,
        radius=2,
        showscale=False,
    )
)

fig.update_layout(
    mapbox_style="carto-darkmatter",
    mapbox_center_lon=133,
    mapbox_center_lat=-25,
    mapbox_zoom=2.5,
)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})

fig.show()
```

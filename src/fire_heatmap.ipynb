```python
import pandas as pd


raw_df = pd.read_csv("../data/gen_2/unprocessed/fire_archive_M-C61_28737.csv")

raw_df
```

```python
confident_df = raw_df[raw_df.confidence > 80]

confident_df
```

```python
required_columns = ["latitude", "longitude", "acq_date"]

required_df = confident_df[required_columns]

required_df
```

```python
rounded_df = required_df.round({"latitude": 2, "longitude": 2})

rounded_df
```

```python
coord_string_df = rounded_df
coord_string_df["coordinates"] = rounded_df.apply(lambda row: f"{row.latitude},{row.longitude}", axis=1)

coord_string_df
```

```python
grouped_df = coord_string_df.groupby(["coordinates"]).size().reset_index(name="count")

grouped_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_df = grouped_df

scaled_df['probability'] = scaler.fit_transform(grouped_df['count'].to_numpy().reshape(-1, 1))
scaled_df = scaled_df.drop(columns=["count"])

scaled_df
```

```python
scaled_df.to_csv("../data/gen_2/processed/fire_coordinate_weights.csv", index=False)
```

```python
import pandas as pd


raw_df = pd.read_csv("../data/gen_2/unprocessed/fire_archive_M-C61_28737.csv")

raw_df
```

```python
confident_df = raw_df[raw_df.confidence > 80]

confident_df
```

```python
required_columns = ["latitude", "longitude", "acq_date"]

required_df = confident_df[required_columns]

required_df
```

```python
rounded_df = required_df.round({"latitude": 1, "longitude": 1})

rounded_df
```

```python
month_df = rounded_df.copy(deep=True)

month_df['month'] = rounded_df['acq_date'].str[5:7]
```

```python
grouped_df = month_df.groupby(["month", "latitude", "longitude"]).size().reset_index(name="count")

grouped_df
```

```python
grouped_df = grouped_df.sample(50000)
```

```python
grouped_df.describe()
```

```python
grouped_df.to_csv("../data/gen_2/processed/fire_coordinate_weights.csv", index=False)
```

```python
import plotly.express as px

# get january
jan_df = grouped_df[grouped_df.month == "12"]

fig = px.density_mapbox(jan_df, lat = 'latitude', lon = 'longitude', z = 'count',
                        radius = 8,
                        center = dict(lat = -20, lon = 140),
                        zoom = 4,
                        mapbox_style = 'open-street-map')

fig.show()
```

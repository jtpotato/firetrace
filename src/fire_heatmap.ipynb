```python
import pandas as pd


raw_df = pd.read_csv("../data/gen_2/unprocessed/fire_archive_M-C61_28737.csv")

raw_df
```

```python
confident_df = raw_df[raw_df.confidence > 80]

confident_df
```

```python
required_columns = ["latitude", "longitude", "acq_date"]

required_df = confident_df[required_columns]

required_df
```

```python
rounded_df = required_df.round({"latitude": 1, "longitude": 1})

rounded_df
```

```python
# coord_string_df = rounded_df
# coord_string_df["coordinates"] = rounded_df.apply(lambda row: f"{row.latitude},{row.longitude}", axis=1)

# coord_string_df
```

```python
grouped_df = rounded_df.groupby(["latitude", "longitude"]).size().reset_index(name="count")

grouped_df
```

```python
grouped_df.describe()
```

```python
grouped_df.to_csv("../data/gen_2/processed/fire_coordinate_weights.csv", index=False)
```

```python
import plotly.express as px

fig = px.density_mapbox(grouped_df, lat = 'latitude', lon = 'longitude', z = 'count',
                        radius = 8,
                        center = dict(lat = -20, lon = 140),
                        zoom = 4,
                        mapbox_style = 'open-street-map')

fig.show()
```
